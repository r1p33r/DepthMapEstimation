# Visualize

```python
import re

def calculate_average(log_file):
    g_gan_values = []
    g_l1_values = []
    d_real_values = []
    d_fake_values = []

    with open(log_file, 'r') as file:
        lines = file.readlines()
        for line in lines:
            match = re.search(r'G_GAN: ([\d.]+) G_L1: ([\d.]+) D_real: ([\d.]+) D_fake: ([\d.]+)', line)
            if match:
                g_gan_values.append(float(match.group(1)))
                g_l1_values.append(float(match.group(2)))
                d_real_values.append(float(match.group(3)))
                d_fake_values.append(float(match.group(4)))

    g_gan_average = sum(g_gan_values) / len(g_gan_values)
    g_l1_average = sum(g_l1_values) / len(g_l1_values)
    d_real_average = sum(d_real_values) / len(d_real_values)
    d_fake_average = sum(d_fake_values) / len(d_fake_values)

    return g_gan_average, g_l1_average, d_real_average, d_fake_average

# Chamando a funo com o nome do arquivo de log
g_gan_avg, g_l1_avg, d_real_avg, d_fake_avg = calculate_average('/media/pc/DATA/Gan-test/pytorch-pix2pix-nyu-grey/checkpoints/pix2pix-nyu/loss_log.txt')

print("G_GAN Average:", g_gan_avg)
print("G_L1 Average:", g_l1_avg)
print("D_real Average:", d_real_avg)
print("D_fake Average:", d_fake_avg)
```

```python
import re
import matplotlib.pyplot as plt

def plot_loss(log_file):
    g_gan_loss = []
    g_l1_loss = []
    d_real_loss = []
    d_fake_loss = []

    with open(log_file, 'r') as file:
        lines = file.readlines()
        for line in lines:
            match = re.search(r'G_GAN: ([\d.]+) G_L1: ([\d.]+) D_real: ([\d.]+) D_fake: ([\d.]+)', line)
            if match:
                g_gan_loss.append(float(match.group(1)))
                g_l1_loss.append(float(match.group(2)))
                d_real_loss.append(float(match.group(3)))
                d_fake_loss.append(float(match.group(4)))

    # Plotting the losses
    epochs = range(1, len(g_gan_loss) + 1)
    plt.plot(epochs, g_gan_loss, label='G_GAN')
    plt.plot(epochs, g_l1_loss, label='G_L1')
    plt.plot(epochs, d_real_loss, label='D_real')
    plt.plot(epochs, d_fake_loss, label='D_fake')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Losses over Epochs')
    plt.show()

# Chamando a funo com o nome do arquivo de log
plot_loss('/media/pc/DATA/Gan-test/pytorch-pix2pix-nyu-grey/checkpoints/pix2pix-nyu/loss_log.txt')
```
